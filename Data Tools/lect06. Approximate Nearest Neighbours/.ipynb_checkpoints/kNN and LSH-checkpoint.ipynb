# Что такое векторный поиск

Поиск похожих объектов в данных — стандартная операция в базах данных, поисковых системах и многих других приложениях. Когда вы работаете с традиционными базами данных, поиск элементов по фиксированным числовым критериям, например, поиск сотрудников в базе данных с зарплатой в определённом диапазоне, выполняется очень просто с помощью SQL.

Но иногда нам приходится отвечать на вопросы вроде: «Какие объекты в нашем ассортименте похожи на то, что искал пользователь?» Поисковые запросы могут быть размытыми и иметь множество вариаций. Например, пользователь может искать что-то общее, например «обувь», «черная обувь», или что-то более конкретное, например «Nike AF-1 LV8».

<img src = "https://www.pinecone.io/_next/image/?url=https%3A%2F%2Fcdn.sanity.io%2Fimages%2Fvr8gru94%2Fproduction%2F2b266c0aa797f3e055a522d562fcaba545193d1b-2113x352.png&w=3840&q=75">

Система должна уметь различать такие запросы и понимать, чем черные туфли отличаются от другой обуви. Для обработки подобных запросов нам нужно представление, которое понимает концептуальное значение объектов. Более того, в таких задачах часто придется работать с данными в огромном количестве, порядка миллиардов объектов.

При работе с данными такого масштаба задача сильно отличается от поиска в традиционных базах данных, содержащих символьные представления объектов. 

С помощью векторного поиска мы можем работать с семантическими представлениями наших данных и быстро находить похожие элементы. В следующих разделах мы обсудим, как именно это работает.

**Что такое векторные представления?**

В машинном обучении мы часто представляем реальные объекты и концепции в виде наборов чисел, также известных как эмбеддинги. Этот метод позволяет нам перевести воспринимаемое нами сходство между объектами в векторное пространство.

Это означает, что когда мы представляем изображения или фрагменты текста в виде векторных представлений, их семантическое сходство определяется тем, насколько близки их векторы в векторном пространстве. Получается, что важная для нас информация —  расстояние между векторами объектов.

Эти векторные представления (эмбеддинги) часто создаются путем обучения нейросетевых моделей в соответствии с входными данными и типом задачи. Word2Vec, GLoVE, USE, T5 и т.д. — список моделей для генерации эмбеддингов из текстовых данных, в то же время для картинок можно использовать CNN-модели или Vision Transformer'ы для создания эмбеддингов.

**Расстояние между векторами**

Как уже говорилось, мы находим сходства между объектами, вычисляя расстояние между их векторами. Можно считать расстояние между векторами, используя разные метрики (выбирая ту, которая лучше всего подходит для нашей задачи).

Некоторые из часто используемых метрик расстояния в машинном обучении — это Евклидово расстояние, расстояние Манхэттена, Косинусное расстояние и расстояние Чебышева.

**Приближенный поиск соседей**

Чтобы снизить вычислительную сложность, связанную с полным перебором, как в k-NN, используется приближенный поиск соседей.

Вместо того чтобы вычислять расстояния до каждого вектора в базе данных, мы получаем «достаточно хорошее приближение» ближайшего соседа. По сути, мы готовы пожертвовать частью точности ради значительного прироста в производительности, что позволяет масштабировать поиск. 

При приближенном поиске ближайших соседей строятся структуры, которые называются индексами, которые сужают область поиска. Кроме того, большинство моделей машинного обучения создают векторы высокой размерности. Приближенный поиск основывается на том факте, что даже если данные представлены векторами большой размерности, реальная информационная нагрузка данных часто не так высока. Он работает как бы с "истинной внутренней размерностью данных". 

Мы разберем разные варианты решения задачи и рассмотрим примеры применения :)

```python
from tqdm import tqdm
import shutil
import urllib.request as request
from contextlib import closing

# Сначала загружаем набор данных Sift1M
with closing(request.urlopen('ftp://ftp.irisa.fr/local/texmex/corpus/sift.tar.gz')) as r:
    # Получаем общий размер файла для прогресс-бара
    file_size = int(r.headers['Content-Length'])
    with open('sift.tar.gz', 'wb') as f:
        with tqdm(total=file_size, unit='B', unit_scale=True, desc='Загрузка Sift1M') as pbar:
            while True:
                chunk = r.read(8192)
                if not chunk:
                    break
                f.write(chunk)
                pbar.update(len(chunk))
```

```python
import tarfile

# the download leaves us with a tar.gz file, we unzip it
tar = tarfile.open('sift.tar.gz', "r:gz")
tar.extractall()
```

```python
import numpy as np

# now define a function to read the fvecs file format of Sift1M dataset
def read_fvecs(fp):
    a = np.fromfile(fp, dtype='int32')
    d = a[0]
    return a.reshape(-1, d + 1)[:, 1:].copy().view('float32')
```

```python
# data we will search through
xb = read_fvecs('./sift/sift_base.fvecs')  # 1M samples
# also get some query vectors to search with
xq = read_fvecs('./sift/sift_query.fvecs')
# take just one query (there are many in sift_learn.fvecs)
xq = xq[0].reshape(1, xq.shape[1])
```

# Locality Sensitive Hashing (LSH) 

LSH – это широко распространенный метод, используемый в приближенном поиске ближайших соседей (ANN). 

Крупные компании, такие как Google, Netflix, Amazon, Spotify, Uber и многие другие, полагаются на поиск по сходству для многих своих ключевых функций.

Amazon использует поиск по сходству для сравнения клиентов, находя новые рекомендации товаров на основе истории покупок наиболее похожих клиентов.

Каждый раз, когда вы используете Google, вы выполняете поиск по сходству между вашим запросом и проиндексированными документами из интернета.

Если Spotify рекомендует хорошую музыку, это потому, что их алгоритмы поиска по сходству успешно сопоставляют вас с другими клиентами с одинаково хорошим (или не очень) музыкальным вкусом.

LSH является одной из первоначальных методик, обеспечивающих высококачественный поиск при сохранении высокой скорости. Разберем этот алгоритм по кирпичикам.

**Сложность поиска**

Представьте набор данных, содержащий миллиарды примеров — как мы можем эффективно их сравнить?

Даже на лучшем оборудовании попарное сравнение всех элементов невозможно. Это дает в лучшем случае сложность $O(n^2)$. Даже при сравнении одного запроса с миллиардами образцов мы все равно получаем сложность $O(n)$.

Нам также необходимо учитывать сложность отдельного вычисления сходства — каждый образец хранится в виде вектора, часто очень высокой размерности — это еще больше увеличивает сложность.

Как можно этого избежать? Возможно ли вообще выполнить поиск со сложностью меньше линейной? Ответ: Да, возможно.

Вместо полного перебора мы можем аппроксимировать и ограничить область поиска только наиболее релевантными векторами.


Рассмотрим векторный индекс. Если бы мы захотели добавить всего один новый вектор и попытаться найти ближайшее соответствие — мы должны сравнить этот вектор с каждым другим вектором в нашей базе данных. Это дает нам линейную временную сложность — что не может масштабироваться для быстрого поиска в больших наборах данных.

Проблема становится еще хуже, если мы хотим сравнить все векторы друг с другом — оптимальный метод сортировки для достижения этого имеет в лучшем случае логарифмически-линейную временную сложность.

Поэтому нам нужен способ сократить количество сравнений. В идеале мы хотим сравнивать только векторы, которые считаем потенциальными соответствиями — или кандидатными парами.

Локально-чувствительное хеширование (LSH) позволяет нам сделать это.

LSH включает в себя различные методы. В этой статье мы рассмотрим традиционный подход, который состоит из нескольких этапов — **Shingling**, **MinHashing** и финальная **banded LSH-функция**.

В своей основе финальная LSH функция позволяет нам сегментировать и хешировать один и тот же образец несколько раз. И если пара векторов после хеширования попала в одно и то же значение хотя бы один раз, мы помечаем пару как кандидата на соответствие.

Это очень похоже на процесс, используемый в словарях Python. У нас есть пара ключ-значение, которую мы помещаем в словарь. Ключ обрабатывается через хеш-функцию словаря и отображается в определенный бакет. Затем мы связываем соответствующее значение с этим бакетом.

<img src = "https://www.pinecone.io/_next/image/?url=https%3A%2F%2Fcdn.sanity.io%2Fimages%2Fvr8gru94%2Fproduction%2Ffaaae9e55d33b08629574243d8456446650df096-1400x787.png&w=3840&q=75">

Однако существует ключевое различие между этим типом хеш-функции и той, что используется в LSH. В словарях наша цель — минимизировать вероятность коллизии.

LSH работает +- противоположным образом. В LSH мы хотим максимизировать коллизии — хотя в идеале – только для похожих входных данных.

<img src = "https://www.pinecone.io/_next/image/?url=https%3A%2F%2Fcdn.sanity.io%2Fimages%2Fvr8gru94%2Fproduction%2F862f88182a796eb16942c47d93ee03ba4cdaee4d-1920x1080.png&w=3840&q=75">

Не существует единого подхода к хешированию в LSH. Действительно, все они разделяют общую логику «помещения похожих образцов в один бакет с помощью хеш-функции», но могут значительно различаться в деталях.

Метод, который мы будем рассматривать далее , можно назвать традиционным подходом, использующим **Shingling**, **MinHashing** и финальная **banded LSH**..

# k-Shingling

k-Shingling - процесс преобразования текстовой строки в набор «шинглов». Перемещаем окно размера k вдоль нашей текстовой строки и «фотографируем» на каждом шаге. Далее собираем все «слепки», чтобы создать наш набор шинглов.

```python
a = "flying fish flew by the space station"
b = "we will not allow you to bring your pet armadillo along"
c = "he figured a few sticks of dynamite were easier than a fishing pole to catch fish"
```

```python
def shingle(text: str, k: int):
    shingle_set = []
    for i in range(len(text) - k+1):
        shingle_set.append(text[i:i+k])
    return set(shingle_set)
```

```python
k = 2

a = shingle(a, k)
b = shingle(b, k)
c = shingle(c, k)
print(a)
```

Следующим шагом мы создаем разреженные векторы (типа Bag of words). Для этого нам сначала нужно объединить все наши наборы, чтобы создать один большой набор, содержащий все шинглы из всех наших наборов — мы называем это словарем (vocab).

Затем мы можем представить каждый исходный документ (или строку) в виде вектора относительно этого словаря. Размерность вектора равна размеру словаря. Если шингл из словаря присутствует в документе, в соответствующей позиции вектора ставится 1, в противном случае — 0.

# MinHashing

**MinHashing** — следующий шаг в нашем процессе, позволяющий преобразовать разреженные векторы в обычные. Предупреждаю заранее — эта часть процесса может показаться запутанной на первый взгляд — но как только вы разберетесь, она окажется очень простой.

У нас есть разреженные векторы, и мы делаем следующее: мы случайным образом генерируем одну minhash-функцию для каждой позиции в нашей сигнатуре (т.е. в плотном векторе).

Таким образом, если мы хотим создать плотный вектор/сигнатуру из 20 чисел — мы используем 20 minhash-функций.

Эти minhash-функции представляют собой просто рандомизированный порядок чисел — и мы считаем "от 1 до конечного числа `len(vocab)`". Поскольку порядок этих чисел был рандомизирован, мы можем обнаружить, что число *1* находится, например, на позиции *57* в нашей рандомизированной минхэш-функции.


## Шаг 1 - сделали vocab

```python
vocab = list(set(a.union(b).union(c)))
print(vocab)
```

## Шаг 2 - сделали minhash

```python
hash_ex = list(range(1, len(vocab)+1))
print(hash_ex) 
```

```python
from random import shuffle

shuffle(hash_ex)
print(hash_ex)
```

Проходим по этому рандомизированному MinHash-вектору (начиная с 1) и сопоставляем индекс каждого значения с эквивалентными значениями в разреженном векторе `a_1hot`. Если мы находим 1 — этот индекс и есть значение нашей сигнатуры.

```python
print(f"7 -> {hash_ex.index(7)}")
```

```python
for i in range(1, 5):
    print(f"{i} -> {hash_ex.index(i)}")
```

```python
a_1hot = [1 if x in a else 0 for x in vocab]
b_1hot = [1 if x in b else 0 for x in vocab]
c_1hot = [1 if x in c else 0 for x in vocab]
print(a_1hot)
```

```python
for i in range(1, len(vocab)+1):
    idx = hash_ex.index(i)
    signature_val = a_1hot[idx]
    print(f"{i} -> {idx} -> {signature_val}")
    if signature_val == 1:
        print('match!')
        break
```

Вот это последнее число – наша сигнатура. Теперь повторяем это дело многократно.

```python
def create_hash_func(size: int):
    hash_ex = list(range(1, len(vocab)+1))
    shuffle(hash_ex)
    return hash_ex

def build_minhash_func(vocab_size: int, nbits: int):
    hashes = []
    for _ in range(nbits):
        hashes.append(create_hash_func(vocab_size))
    return hashes

# делаем 20 minhash векторов
minhash_func = build_minhash_func(len(vocab), 20)
```

```python
def create_hash(vector: list):
    # use this function for creating our signatures (eg the matching)
    signature = []
    for func in minhash_func:
        for i in range(1, len(vocab)+1):
            idx = func.index(i)
            signature_val = vector[idx]
            if signature_val == 1:
                signature.append(idx)
                break
    return signature
```

```python
a_sig = create_hash(a_1hot)
b_sig = create_hash(b_1hot)
c_sig = create_hash(c_1hot)

print(a_sig)
print(b_sig)
```

Проверим, что там с пересечениями (посчитаем Жаккара aka intersection over union)

```python
def jaccard(a: set, b: set):
    return len(a.intersection(b)) / len(a.union(b))

jaccard(a, b), jaccard(set(a_sig), set(b_sig))
```

```python
jaccard(a, c), jaccard(set(a_sig), set(c_sig))
```

```python
jaccard(b, c), jaccard(set(b_sig), set(c_sig))
```

## Шаг 3 - сделали banding

```python
def split_vector(signature, b):
    assert len(signature) % b == 0
    r = int(len(signature) / b)
    subvecs = []
    for i in range(0, len(signature), r):
        subvecs.append(signature[i : i+r])
    return subvecs
```

```python
import requests
import pandas as pd
import io

url = "https://raw.githubusercontent.com/brmson/dataset-sts/master/data/sts/sick2014/SICK_train.txt"

text = requests.get(url).text

data = pd.read_csv(io.StringIO(text), sep='\t')
data.head()
```

```python
sentences = data['sentence_A'].tolist()
sentences[:3]
```

```python
def build_shingles(sentence: str, k: int):
    shingles = []
    for i in range(len(sentence) - k):
        shingles.append(sentence[i:i+k])
    return set(shingles)

def build_vocab(shingle_sets: list):
    # convert list of shingle sets into single set
    full_set = {item for set_ in shingle_sets for item in set_}
    vocab = {}
    for i, shingle in enumerate(list(full_set)):
        vocab[shingle] = i
    return vocab

def one_hot(shingles: set, vocab: dict):
    vec = np.zeros(len(vocab))
    for shingle in shingles:
        idx = vocab[shingle]
        vec[idx] = 1
    return vec
```

```python
k = 8  # shingle size


shingles = []
for sentence in sentences:
    shingles.append(build_shingles(sentence, k))


vocab = build_vocab(shingles)


shingles_1hot = []
for shingle_set in shingles:
    shingles_1hot.append(one_hot(shingle_set, vocab))

shingles_1hot = np.stack(shingles_1hot)
shingles_1hot.shape
```

# MinHash

```python
def minhash_arr(vocab: dict, resolution: int):
    length = len(vocab.keys())
    arr = np.zeros((resolution, length))
    for i in range(resolution):
        permutation = np.random.permutation(len(vocab)) + 1
        arr[i, :] = permutation.copy()
    return arr.astype(int)

def get_signature(minhash, vector):
    # get index locations of every 1 value in vector
    idx = np.nonzero(vector)[0].tolist()
    # use index locations to pull only +ve positions in minhash
    shingles = minhash[:, idx]
    # find minimum value in each hash vector
    signature = np.min(shingles, axis=1)
    return signature
```

```python
arr = minhash_arr(vocab, 100)

signatures = []

for vector in shingles_1hot:
    signatures.append(get_signature(arr, vector))

# merge signatures into single array
signatures = np.stack(signatures)
signatures.shape
```

```python
signatures[0]
```

```python
from itertools import combinations

class LSH:
    buckets = []
    counter = 0
    def __init__(self, b):
        self.b = b
        for i in range(b):
            self.buckets.append({})

    def make_subvecs(self, signature):
        l = len(signature)
        assert l % self.b == 0
        r = int(l / self.b)
        # break signature into subvectors
        subvecs = []
        for i in range(0, l, r):
            subvecs.append(signature[i:i+r])
        return np.stack(subvecs)
    
    def add_hash(self, signature):
        subvecs = self.make_subvecs(signature).astype(str)
        for i, subvec in enumerate(subvecs):
            subvec = ','.join(subvec)
            if subvec not in self.buckets[i].keys():
                self.buckets[i][subvec] = []
            self.buckets[i][subvec].append(self.counter)
        self.counter += 1

    def check_candidates(self):
        candidates = []
        for bucket_band in self.buckets:
            keys = bucket_band.keys()
            for bucket in keys:
                hits = bucket_band[bucket]
                if len(hits) > 1:
                    candidates.extend(combinations(hits, 2))
        return set(candidates)
```

```python
b = 20

lsh = LSH(b)

for signature in signatures:
    lsh.add_hash(signature)
```

```python
candidate_pairs = lsh.check_candidates()
len(candidate_pairs)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def jaccard(a: set, b: set):
    return len(a.intersection(b)) / len(a.union(b))
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

pairs_data = {
    'x': [],
    'y': [],
    'jaccard': [],
    'cosine': [],
    'candidate': []
}

data_len = shingles_1hot.shape[0]
chosen = set()
# take random sample of pairs
sample_size = 50_000
for _ in range(sample_size):
    x, y = np.random.choice(data_len, 2)
    if x == y or (x, y) in chosen: 
        continue
    chosen.add((x, y))
    vector_x = signatures[x]
    vector_y = signatures[y]
    candidate = 1 if (x, y) in candidate_pairs else 0
    cosine = cosine_similarity([vector_x], [vector_y])[0][0]
    
    # Добавляем данные в словарь
    pairs_data['x'].append(x)
    pairs_data['y'].append(y)
    pairs_data['jaccard'].append(jaccard(set(vector_x), set(vector_y)))
    pairs_data['cosine'].append(cosine)
    pairs_data['candidate'].append(candidate)

# Создаем DataFrame из собранных данных
pairs = pd.DataFrame(pairs_data)

cos_min = pairs['cosine'].min()
cos_max = pairs['cosine'].max()
pairs['cosine_norm'] = (pairs['cosine'] - cos_min) / (cos_max - cos_min)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(data=pairs, x='cosine', y='candidate', alpha=0.5)
```

```python
def probability(s, r, b):
    # s: similarity
    # r: rows (per band)
    # b: number of bands
    return 1 - (1 - s**r)**b

def normalize(x, x_min, x_max):
    return (x - x_min) / (x_max - x_min)
```

```python
b = 20
r = int(100 / b)
s_scores = np.arange(0.01, 1, 0.01)
P_scores = [probability(s, r, b) for s in s_scores]

sns.lineplot(x=s_scores, y=P_scores)
sns.scatterplot(data=pairs, x='cosine', y='candidate', alpha=0.1, color='k')
```

```python
b = 20
r = int(100 / b)
s_scores = np.arange(0.01, 1, 0.01)
P_scores = [probability(s, r, b) for s in s_scores]

sns.lineplot(x=s_scores, y=P_scores)
sns.scatterplot(data=pairs, x='cosine_norm', y='candidate', alpha=0.1, color='k')
```

```python
probs_data = []

for b in [100, 50, 25, 20, 10, 5, 2]:
    r = int(100 / b)
    s_scores = np.arange(0.01, 1, 0.01)
    P_scores = [probability(s, r, b) for s in s_scores]
    
    # Собираем данные в список
    for s, P in zip(s_scores, P_scores):
        probs_data.append({
            'P': P,
            's': s,
            'b': str(b)
        })

# Создаем DataFrame из собранных данных
probs = pd.DataFrame(probs_data)

sns.lineplot(data=probs, x='s', y='P', hue='b')
```

# Определения

Пусть $A,B$ — конечные множества, $U=A\cup B$, $I=A\cap B$.

**Cхожесть по Жаккару**: $J(A,B)=\dfrac{|I|}{|U|}$, 

**Расстояние Жаккара**: $d_J(A,B)=1-J(A,B)$.

Пусть $\pi$ — случайная перестановка элементов.

**MinHash по $\pi$**: $h_\pi(S)=\arg\min_{x\in S}\pi(x)$ — элемент множества $S$ с минимальным рангом в порядке $\pi$.

---

# классический факт про MinHash

$\Pr\big[h_\pi(A)=h_\pi(B)\big]=J(A,B)=\frac{|A\cap B|}{|A\cup B|}.$

**Доказательство.** Рассмотрим объединение $U$. В случайной перестановке $\pi$ каждый $u\in U$ **равновероятно** может оказаться минимальным (первым) элементом $U$; 
значит

$\Pr(\text{минимум в }U\text{ равен }u)=\frac{1}{|U|}\quad \text{для каждого }u\in U.$

Событие $h_\pi(A)=h_\pi(B)$ эквивалентно тому, что **минимум по $U$** лежит в **пересечении $I$** (тогда этот элемент принадлежит и $A$, и $B$). 

Поэтому
$\Pr\big[h_\pi(A)=h_\pi(B)\big]
=\sum_{u\in I}\Pr{\text{минимум}=u}
=\sum_{u\in I}\frac{1}{|U|}
=\frac{|I|}{|U|}
=J(A,B).$
Что и требовалось. $\square$

---

# Следствие: сохранение Жаккарова **расстояния**

Построим подпись длины $k$: возьмём независимые перестановки $\pi_1,\dots,\pi_k$ и значения $h_{\pi_i}(A),h_{\pi_i}(B)$.

Введём индикаторы совпадения
$X_i=\mathbf{1}({h_{\pi_i}(A)=h_{\pi_i}(B)}),\qquad i=1,\dots,k.$

По теореме выше $X_i\sim\text{Bernoulli}(p)$ с $p=J(A,B)$ и независимы. 

Тогда
$\widehat{J}=\frac{1}{k}\sum_{i=1}^k X_i,\qquad
\mathbb{E}[\widehat{J}]=J(A,B),\qquad
\mathrm{Var}(\widehat{J})=\frac{p(1-p)}{k}.$

Оценка **Жаккарова расстояния** из подписи:
$\widehat{d}_J=1-\widehat{J}.$
Тогда $\mathbb{E}[\widehat{d}_J]=1-\mathbb{E}[\widehat{J}]=d_J(A,B)$.

То есть MinHash **сохраняет Жаккарово расстояние в среднем** (а при большом $k$ — с высокой точностью).

---

# Неравенство Хёффдинга и концентрация

Поскольку $X_i\in[0,1]$ независимы, к $\widehat{J}$ применимо неравенство Хёффдинга.

**Неравенство Хёффдинга.** Для независимых $Y_i\in[a_i,b_i]$,

$\Pr\left(\frac{1}{n}\sum_{i=1}^n Y_i - \mathbb{E}\left[\frac{1}{n}\sum Y_i\right]\ge \varepsilon\right)
\le \exp\left(-\frac{2n^2\varepsilon^2}{\sum_{i=1}^n (b_i-a_i)^2}\right).$

В симметричной форме:
$\Pr\left(\left|\frac{1}{n}\sum Y_i - \mathbb{E}\left[\frac{1}{n}\sum Y_i\right]\right|\ge \varepsilon\right)
\le 2\exp\left(-\frac{2n^2\varepsilon^2}{\sum (b_i-a_i)^2}\right).$

Подставим $Y_i=X_i$, $a_i=0$, $b_i=1$, $n=k$. 

Получаем
$\Pr\big(|\widehat{J}-J|\ge \varepsilon\big)\le 2\exp(-2k\varepsilon^2),$

и, эквивалентно, для расстояния

$\Pr\big(|\widehat{d}_J-d_J|\ge \varepsilon\big)\le 2\exp(-2k\varepsilon^2).$

Отсюда **размер подписи** для требуемой точности: 

чтобы
$\Pr\big(|\widehat{d}_J-d_J|\le \varepsilon\big)\ge 1-\delta,$

достаточно взять

$k \ge \frac{1}{2\varepsilon^2}\ln\frac{2}{\delta}.$

---

## Тонкость про «идеальность»

Доказательство предполагает **истинно случайную перестановку**. Равенство выполняется с очень малой систематической ошибкой, а концентрация по Хёффдингу остаётся применимой при разумном предположении независимости хешей.
